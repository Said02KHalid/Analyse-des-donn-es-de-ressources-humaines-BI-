```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv('HR_SAMIR_SAID.csv')
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
df.dropna(thresh=len(df.columns) * 0.6, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.dtypes)
```

```python
print(df['ManagerID'].isnull().sum())
```

```python
df['ManagerID'] = df['ManagerID'].fillna(0)

```

```python
df['ManagerID'] = df['ManagerID'].astype(int)
```

```python
print(df['ManagerID'].dtypes)
```

```python
print(df.isnull().sum())
```

```python
df.drop(columns=['DateofTermination'], inplace=True)
```

```python
df['Employee_Name'] = df['Employee_Name'].str.replace(',', ' ')
```

```python
print(df.dtypes)
```

```python
# Assuming df is your existing DataFrame
def convert_object_columns_to_unicode(dataframe):
    # Iterate over each column in the DataFrame
    for column in dataframe.select_dtypes(include=['object']).columns:
        # Convert the column to string (Unicode) type
        dataframe[column] = dataframe[column].astype(str)
    return dataframe

# Example usage:
# df = pd.read_csv('your_data.csv')  # Load your DataFrame
df = convert_object_columns_to_unicode(df)

# Verify the conversion
print(df.dtypes)
```
